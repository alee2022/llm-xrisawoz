```python
import json
```

```python
langs = ["en", "ko", "zh", "fr", "hi", "enhi"]
```

```python
for lang in langs:
    filename = "results/ep20_newline/result_" + lang + ".json"
    with open(filename) as json_file:
        results = json.load(json_file)
    outname = "pred_files/pred_" + lang + ".tsv"
    out_DST = "pred_files/pred_" + lang + "_DST.tsv"
    out_API = "pred_files/pred" + lang + "_API.tsv"
    out_DA = "pred_files/pred" + lang + "_DA.tsv"
    out_RG = "pred_files/pred" + lang + "_RG.tsv"
    with open(outname, "w") as f:
        with open(out_DST, "w") as f_dst:
            with open(out_API, "w") as f_api:
                with open(out_DA, "w") as f_da:
                    with open(out_RG, "w") as f_rg:
                        for i in range(len(results)):
                            idx = i
                            pred = results[i]["prediction"]
                            gold = results[i]["output"]
                            inp = results[i]["input"]
                            print("{idx}\t{pred}\t{gold}\t{inp}".format(idx=idx, pred=pred, gold=gold, inp=inp), file=f)
                            if inp[:3] == "DST":
                                print("{idx}\t{pred}\t{gold}\t{inp}".format(idx=idx, pred=pred, gold=gold, inp=inp), file=f_dst)
                            if inp[:3] == "API":
                                print("{idx}\t{pred}\t{gold}\t{inp}".format(idx=idx, pred=pred, gold=gold, inp=inp), file=f_api)
                            if inp[:2] == "DA":
                                print("{idx}\t{pred}\t{gold}\t{inp}".format(idx=idx, pred=pred, gold=gold, inp=inp), file=f_da)
                            if inp[:2] == "RG":
                                print("{idx}\t{pred}\t{gold}\t{inp}".format(idx=idx, pred=pred, gold=gold, inp=inp), file=f_rg)
```
